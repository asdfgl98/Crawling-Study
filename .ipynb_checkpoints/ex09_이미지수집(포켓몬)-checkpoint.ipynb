### 포켓몬 페이지에서 이미지 데이터 수집하기
- 이미지를 다운로드 받아서 저장할 폴더생성 > 자동화
- 이미지 저장

```python
# 파일 시스템을 위한 라이브러리 / 파일 폴더 생성할 때 사용하는 라이브러리
import os # Operating System
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup as bs
import pandas as pd
# 이미지의 경로를 파일로 변경해주는 라이브러리
from urllib.request import urlretrieve
```

- 바탕화면에 '포켓몬도감' 폴더생성

```python
# os 라이브러리를 활용해서 폴더 생성
# \ 를 하나만 사용하면 이스케이프 코드로 동작하기때문에
# \ 자체를 사용하고 싶다면 \ 앞에 하나를 더 붙여주어야한다
loc = 'C:\\Users\\gjaischool1\\Desktop\\포켓몬도감'
if not os.path.isdir(loc):
    print('폴더생성')
    os.mkdir(loc)
else:
    print('폴더가 이미 존재합니다.')
```

```python
# 포켓몬 사이트 열기
driver = wb.Chrome()
driver.get('https://www.pokemonkorea.co.kr/pokedex')
```

```python
# img 태그의 속성 값 추출을 위해 bs 객체화
# bs(페이지정보, 파싱방법)
soup = bs(driver.page_source, 'lxml')
soup
```

```python
# 포켓몬 태그 추출
img = soup.select('img.img-fluid')
img
```

```python
# .text 요소에서 콘텐츠를 추출할때 사용
# img 태그는 홀태그로써 콘텐츠를 가지지 않는다
# 요소에서 속성을 추출해야한다
```

- 수집한 요소에서 콘텐츠가 아닌 속성 추출하기

```python
# 요소['속성명']
img[0]['src']
# 출력된 정보 : 이미지의 경로, 주소
```

```python
# 파일의 경로를 활용하여 이미지 파일로 저장
# urlretrieve(이미지의 경로, '저장할 위치\\파일명.확장자')
urlretrieve(img[0]['src'], loc+'\\이상해씨.jpg')
```

```python
len(img)
```

```python
# 순수한 src 주소를 담을 리스트 생성
img_list = []
for i in img:
    img_list.append(i['src'])
img_list
```

```python
for i in range(len(img_list)):
    urlretrieve(img_list[i], loc+f'\\포켓몬{i}.jpg')
```

### 화면에서 자동 스크롤 동작 시키기

```python
# 키보드의 end 키를 통해서 페이지 스크롤 내리기
body = driver.find_element(By.TAG_NAME, 'body')
for i in range(20):
    time.sleep(0.5)
    body.send_keys(Keys.END)
```

### 코드 합치기

```python
# os 라이브러리를 활용해서 폴더 생성
# \ 를 하나만 사용하면 이스케이프 코드로 동작하기때문에
# \ 자체를 사용하고 싶다면 \ 앞에 하나를 더 붙여주어야한다
loc = 'C:\\Users\\gjaischool1\\Desktop\\포켓몬도감'
if not os.path.isdir(loc):
    print('폴더생성')
    os.mkdir(loc)
else:
    print('폴더가 이미 존재합니다.')
    
# 포켓몬 사이트 열기
driver = wb.Chrome()
driver.get('https://www.pokemonkorea.co.kr/pokedex')

# 키보드의 end 키를 통해서 페이지 스크롤 내리기
body = driver.find_element(By.TAG_NAME, 'body')
for i in range(20):
    time.sleep(0.5)
    body.send_keys(Keys.END)
    
    
# img 태그의 속성 값 추출을 위해 bs 객체화
# bs(페이지정보, 파싱방법)
soup = bs(driver.page_source, 'lxml')

# 포켓몬 태그 추출
img = soup.select('img.img-fluid')

# 순수한 src 주소를 담을 리스트 생성
img_list = []
for i in img:
    img_list.append(i['src'])

# 이미지 저장
for i in range(len(img_list)):
    urlretrieve(img_list[i], loc+f'\\포켓몬{i}.jpg')
```

### 스크롤 끝까지 내리는 코드

```python
# 스크롤 높이 가져옴
last_height = driver.execute_script("return document.body.scrollHeight")
sleep_time = 1
while True:
    # 끝까지 스크롤 내리기
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 대기
    time.sleep(sleep_time)

    # 스크롤 내린 후 스크롤 높이 다시 가져옴
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```
