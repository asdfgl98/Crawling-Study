```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 1. req를 활용해서 네이버 환율사이트 요청
# * 응답코드를 확인하고, 변수에 담자.
url = 'https://finance.naver.com/marketindex/index.naver'
res = req.get(url)
res
```

```python
# 2. bs를 통해서 컴퓨터가 이해할 수 있는 html 형태의 데이터로 변형
soup = bs(res.text, 'lxml')
soup.select('iframe')
```

```python
# 3. 환율정보 중 국가명을 수집
soup.select('tbody>tr>a')
```
